# HYPOTHESIS: More government spending on education will result in an increase in lower secondary education completion rates among both sexes.

### Data Sets Used:
    SE.SEC.CMPT.LO.ZS - Lower secondary completion rate, total (% of relevant age group)
    SE.XPD.TOTL.GD.ZS - Government expenditure on education, total (% of GDP)

```python
import pandas as pd

COMPLETE_RATE = "SE.SEC.CMPT.LO.ZS"

GOV_SPENDING = "SE.XPD.TOTL.GD.ZS"
```

## Reads the CSV file into a DataFrame.

```python
data = pd.read_csv("../../data/databank_education_2018.csv")
```

## Outputs the DataFrame. (for personal reference)

```python
data
```

## Prints the shape of the DataFrame.

```python
data.shape
```

## Prints the first 10 rows in DataFrame.

```python
data.head(10)
```

## Isolates three desired columns from the DataFrame.

```python
subset_data = data[["Country", COMPLETE_RATE, GOV_SPENDING]]
```

## Prints the subset data. (for personal reference)

```python
subset_data
```

## Prints the first 10 rows of the subset data.

```python
subset_data.head(10)
```

## Creates four booleans for which to filter out the rows with.

```python
notna_completerate = subset_data[COMPLETE_RATE].notna()
```

```python
notna_govspending = subset_data[GOV_SPENDING].notna()
```

```python
nonapp_completerate = subset_data[COMPLETE_RATE] != "N/A"
```

```python
notapp_govspending = subset_data[GOV_SPENDING] != "N/A"
```

## Prints rows with only complete data. Filters out all rows with NaN data.

```python
filtered_data = subset_data[notna_completerate & notna_govspending & nonapp_completerate & notapp_govspending]

filtered_data
```

## Prints the shape of filtered out DataFrame.

```python
filtered_data.shape
```

## Prints only first 10 rows of filtered DataFrame.

```python
filtered_data.head(10)
```

## Creates a new DataFrame with data of government spending in descending order.

```python
descending_data = filtered_data[GOV_SPENDING].sort_values(ascending=False)

descending_data
```

## Prints first 10 rows of descending government spending data.

```python
descending_data.head(10)
```

## Plots the filtered data.

```python
plot_data = filtered_data[[COMPLETE_RATE, GOV_SPENDING]]

import matplotlib.pyplot as plt
```

```python
plot_data.plot.scatter(x=GOV_SPENDING, y=COMPLETE_RATE)
plt.xlabel("% of GDP Government Spending")
plt.ylabel("% of relevant age group who completed secondary education", size= 8)
plt.title("Government Spendings vs Secondary Education Completion")
```

## Finds the correlation between the two sets of data.

```python
govspend_data = filtered_data[GOV_SPENDING]

completerate_data = filtered_data[COMPLETE_RATE]

govspend_data.corr(completerate_data)
```

## Plot of the data with a linear regression line.

```python
import numpy as np
```

```python
x = np.array([govspend_data])[0]
y = np.array([completerate_data])[0]
```

```python
plt.plot(x, y, 'o')

m, b = np.polyfit(x, y, 1)

plt.plot(x, m*x+b)

plt.xlabel("% of GDP Government Spending")
plt.ylabel("% of relevant age group who completed secondary education", size= 8)
plt.title("Government Spendings vs Secondary Education Completion")
```

# CONCLUSION: The data did not create a clear enough correlation to say that the hypothesis was correct. There appears to be a very small correlation between the amount of spending and the completion of secondary education given that the value calculated was only 0.0395 which is very close to 0. Therefore, since the connection was not clear enough to confidently say there is a correlation the results contradict the hypothesis and say that there is likely little to no connection between the two.
